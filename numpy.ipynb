# 🔢 1. NumPy

## 📖 What is NumPy?

NumPy stands for **Numerical Python**.

It is the **foundation of Data Science and AI**.

It helps Python perform mathematical calculations **much faster** than normal Python lists.

---

### 👨‍💻 Created By

Travis Oliphant

📅 Released in **2005**

---

### 🎯 Why Do We Need NumPy?

Suppose you have marks of **10 lakh students**.

Python Lists 😓 → Slow

NumPy Arrays 🚀 → Fast

---
### ⭐ Features

✔ Fast

✔ Memory Efficient

✔ Multidimensional Arrays

✔ Mathematical Functions

✔ Used in AI

✔ Used in Data Science

---

### 🌎 Applications

🤖 Artificial Intelligence

📷 Image Processing

📊 Data Analysis

🌦 Weather Forecast

🏥 Healthcare

🚗 Self Driving Cars

`````python
import numpy as np
import matplotlib.pyplot as plt

# Create Array
arr = np.array([10,20,30,40,50])

print("Array :",arr)
print("Sum :",np.sum(arr))
print("Mean :",np.mean(arr))
print("Maximum :",np.max(arr))

plt.figure(figsize=(5,3))
plt.bar(range(len(arr)),arr)
plt.title("NumPy Array Visualization")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
`````
