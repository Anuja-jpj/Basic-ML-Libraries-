# 📊 2. Pandas

## 📖 What is Pandas?

Pandas is a Python library used to **store, organize and analyze data**.

Think of Pandas like **Microsoft Excel inside Python**.

Instead of working manually with thousands of rows, Pandas can do it in seconds.

---

Wes McKinney

📅 Released in **2008**

📄 Series → One Column

📋 DataFrame → Complete Table

### Why Pandas?

Without Pandas

❌ Difficult Data Analysis

❌ Large Code

With Pandas

✔ Clean Data

✔ Analyze Data

✔ Easy Filtering

✔ Easy Calculations

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Name":["Aman","Priya","Rohit","Sneha"],
    "Marks":[85,92,78,90]
}

df = pd.DataFrame(data)

print(df)

print("\nAverage Marks :",df["Marks"].mean())

plt.figure(figsize=(5,3))
plt.bar(df["Name"],df["Marks"])
plt.title("Student Marks using Pandas")
plt.show()
```

# 🔍 NumPy vs Pandas

| NumPy 🔢 | Pandas 📊 |
|-----------|------------|
| Works with Arrays | Works with Tables |
| Faster Calculations | Easier Data Analysis |
| Mathematical Operations | Data Cleaning |
| Foundation of AI | Data Science |
