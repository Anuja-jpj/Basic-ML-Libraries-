# ⚡ 7. Keras

## 📖 What is Keras?

Keras is a high-level Python library used to build **Artificial Neural Networks** easily.

It works on top of TensorFlow.

Think of it like:

TensorFlow 🏗️ = Building materials

Keras 🛠️ = Easy tools to build the house

# 🎯 Why Do We Need Keras?

Building neural networks directly in TensorFlow can require many lines of code.

✅ Faster

```python
# Simple Neural Network using Keras

import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])


model.compile(
    optimizer="adam",
    loss="binary_crossentropy"
)


model.summary()
```
